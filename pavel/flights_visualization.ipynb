```python
from great_tables import GT, style, loc


# Define a color map for airports
airport_colors = {
    'Miami': "#FFBF00",  # Gold
    'New Orleans': '#87CEEB',  # Sky Blue
    'Mobile Regional Airoprt': '#90EE90',  # Light Green
}

# Function that returns color for a row based on airport
def airport_color_map(table):
    return table['airport'].map(airport_colors)

# Create the table
gt = GT(table)

# Apply background color to airport column cells according to airport_colors mapping
gt = gt.tab_style(
    style=style.fill(color=airport_color_map),
    locations=loc.body(columns=['airport'])
)

# Optionally add header styling and title
gt = gt.tab_style(
    style=style.text(weight='bold'),
    locations=loc.header()
).tab_header(title="Flight Summary by Airport")

gt
```
