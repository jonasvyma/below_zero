```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# safer path string
xls_path = r"C:\Users\rizza\Desktop\gaussian_peppers\dbt_meteostat\below_zero\simone\weather data.xlsx"

# load all sheets
xls = pd.ExcelFile(xls_path)
dfs = [pd.read_excel(xls, sheet_name=s) for s in xls.sheet_names]

# combine into one dataframe
df = pd.concat(dfs, ignore_index=True)

# build a proper datetime column (choose your year, e.g. 2005)
YEAR = 2005
df["date"] = pd.to_datetime({
    "year": YEAR,
    "month": df["month"],
    "day": df["day"]
})

print(df.head())


```

```python
# define date window
start = pd.to_datetime("2005-08-10")
end   = pd.to_datetime("2005-09-10")

# filter for that period
period_df = df[(df["date"] >= start) & (df["date"] <= end)]

# plot
sns.set_theme()
plt.figure(figsize=(12,6))

sns.lineplot(
    data=period_df,
    x="date",
    y="wind_speed",
    hue="airport"
)

# vertical line at 23 August 2005
plt.axvline(pd.to_datetime("2005-08-23"), color="red", linestyle="--", linewidth=2, label="23 Aug 2005")

plt.title("Wind Speed â€” 10 Aug to 10 Sep 2005")
plt.xlabel("Date")
plt.ylabel("Wind Speed")
plt.legend(title="Airport")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('prep_weather_hourly_202509151140.csv')
YEAR = 2005
df["date"] = pd.to_datetime({
    "year": YEAR,
    "month": df["month"],
    "day": df["day"]
})

print(df.head())
```

```python
# define narrower date window
start = pd.to_datetime("2005-08-20")
end   = pd.to_datetime("2005-08-31")

# filter
aug_window = df[(df["date"] >= start) & (df["date"] <= end)]

# plot
plt.figure(figsize=(12,6))
sns.lineplot(
    data=aug_window,
    x="date",
    y="precipitation_mm",
    hue="airport_code"        # make sure this matches your column name
)

# highlight 23 Aug
plt.axvline(pd.to_datetime("2005-08-23"), color="red", linestyle="--", linewidth=2, label="23 Aug 2005")

plt.title("Precipitation â€” 20 to 31 Aug 2005")
plt.xlabel("Date")
plt.ylabel("Precipitation (mm)")
plt.legend(title="Airport")
plt.tight_layout()
plt.show()

```
