```python
from datetime import datetime
from meteostat import Hourly
import pandas as pd

start = datetime(2005, 7, 23)
end   = datetime(2005, 9, 23) 
loc   = '72202'                

data = Hourly(loc, start, end, timezone='America/New_York').fetch()
data.reset_index().to_csv("miami_2months_hourly.csv", index=False)
print("Saved -> miami_MIA_2005-08-23_hourly.csv")

```

```python
weather_daily_df = pd.DataFrame(data)
weather_daily_df
```
